```python
# Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# File names and labels
file_names = ["50_data.csv", "60_data.csv", "70_data.csv", "80_data.csv", "90_data.csv", "100_data.csv"]
labels = ["50_data", "60_data", "70_data", "80_data", "90_data", "100_data"]

# Initialize storage for combined data
dataset_averages = {label: {} for label in labels}
dataset_maxes = {label: {} for label in labels}
dataset_minims = {label: {} for label in labels}

# Colormap for gradient
cmap = cm.get_cmap('coolwarm', len(file_names))  # Change 'coolwarm' to another colormap if desired
colors = [cmap(i) for i in range(len(file_names))]
```

```python
# Process each file
for file_name, label in zip(file_names, labels):
    # Load the data
    data = pd.read_csv(file_name)
    
    # Detect the maximum 'n' from column names
    n_alpha = max([int(col.split('_')[1]) for col in data.columns if col.startswith('alpha_')])
    n_a = max([int(col.split('_')[1]) for col in data.columns if col.startswith('a_')])
    n_b = max([int(col.split('_')[1]) for col in data.columns if col.startswith('b_')])
    n_r = max([int(col.split('_')[1]) for col in data.columns if col.startswith('r_')])
    
    # Calculate averages for alpha, a, b, r
    data['avg_alpha'] = data[[f'alpha_{i}' for i in range(n_alpha + 1)]].mean(axis=1)
    data['avg_a'] = data[[f'a_{i}' for i in range(n_a + 1)]].mean(axis=1)
    data['avg_b'] = data[[f'b_{i}' for i in range(n_b + 1)]].mean(axis=1)
    data['avg_r'] = data[[f'r_{i}' for i in range(n_r + 1)]].mean(axis=1)
    
    # Store averages and loss for plotting
    dataset_averages[label]['epoch'] = data['epoch']
    dataset_averages[label]['loss'] = data['loss']
    dataset_averages[label]['avg_alpha'] = data['avg_alpha']
    dataset_averages[label]['avg_a'] = data['avg_a']
    dataset_averages[label]['avg_b'] = data['avg_b']
    dataset_averages[label]['avg_r'] = data['avg_r']



```

```python
# Plot loss for each dataset
plt.figure(figsize=(10, 6))
for label, color in zip(labels, colors):
    plt.plot(dataset_averages[label]['epoch'], dataset_averages[label]['loss'], label=f"Loss ({label})", color=color)
plt.title('Loss Across Datasets')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot avg_alpha for each dataset
plt.figure(figsize=(10, 6))
for label, color in zip(labels, colors):
    plt.plot(dataset_averages[label]['epoch'], dataset_averages[label]['avg_alpha'], label=f"Alpha ({label})", color=color)
plt.title('Average Alpha Across Datasets')
plt.xlabel('Epoch')
plt.ylabel('Average Alpha')
plt.legend()
plt.show()

# Plot avg_a for each dataset
plt.figure(figsize=(10, 6))
for label, color in zip(labels, colors):
    plt.plot(dataset_averages[label]['epoch'], dataset_averages[label]['avg_a'], label=f"A ({label})", color=color)
plt.title('Average A Across Datasets')
plt.xlabel('Epoch')
plt.ylabel('Average A')
plt.legend()
plt.show()

# Plot avg_b for each dataset
plt.figure(figsize=(10, 6))
for label, color in zip(labels, colors):
    plt.plot(dataset_averages[label]['epoch'], dataset_averages[label]['avg_b'], label=f"B ({label})", color=color)
plt.title('Average B Across Datasets')
plt.xlabel('Epoch')
plt.ylabel('Average B')
plt.legend()
plt.show()

# Plot avg_r for each dataset
plt.figure(figsize=(10, 6))
for label, color in zip(labels, colors):
    plt.plot(dataset_averages[label]['epoch'], dataset_averages[label]['avg_r'], label=f"R ({label})", color=color)
plt.title('Average R Across Datasets')
plt.xlabel('Epoch')
plt.ylabel('Average R')
plt.legend()
plt.show()
```
